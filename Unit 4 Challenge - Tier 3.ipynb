```python
properties_start_prices = properties[properties['Date']=='1995-01-01']
properties_end_prices = properties[properties['Date']=='2021-05-01']
properties_price_change_startend = properties_end_prices.merge(properties_start_prices, on='Borough')
properties_price_change_startend['price_change'] = properties_price_change_startend['Avg House Price_x'] - properties_price_change_startend['Avg House Price_y']
#properties_price_change_startend.sort_values('price_change', ascending=False)
properties_price_change_startend['percent_increase'] = ((properties_price_change_startend['Avg House Price_x'] - properties_price_change_startend['Avg House Price_y']) / properties_price_change_startend['Avg House Price_y']) * 100
properties_price_change_startend.sort_values('percent_increase', ascending=False)
```
